## OpenCV - Image processing

```python
# Install openCV

#! pip install opencv-python
```

```python
import cv2
```

```python
# Load the image using cv2
foto = cv2.imread(r'/content/jorja smith.jpeg')
foto
```

```python
type(foto)
```

```python
# Dimensionality
foto.shape
```

```python
# Visaulize the image
# Opencv uses a different color channel called BGR (Blue, Green, Red) instead of RGB (Red, Green, Blue)
plt.imshow(foto);
```

```python
# Convert from BGR to RGB
foto_c = cv2.cvtColor(foto, cv2.COLOR_BGR2RGB)
plt.imshow(foto_c)



```

```python
# Resizing the image (width, height)
r_foto = cv2.resize(foto_c, (125, 100))
plt.imshow(r_foto);
```

```python
# Flipping the image 
# 0 flips the picture horizontally
ffoto = cv2.flip(foto_c, 0)
plt.imshow(ffoto);
```

```python
# Flipping the image 
# -1 flips the picture horizontally
ffoto = cv2.flip(foto_c, -1)
plt.imshow(ffoto);
```

#### Drawing on Images

```python
# Create a blank Canvas
blank = np.zeros(shape=(1000, 1000,3), dtype=np.int16)
```

```python
np.zeros(10)
```

```python
blank
```

```python
plt.imshow(blank)
```

```python
# Drawing a rectangle
cblank = blank.copy()
```

```python
# PT1 = Top left corner
cv2.rectangle(cblank, pt1=(200,200), pt2=(400,400), color=(255,0,0), thickness=5)
plt.imshow(cblank)
```

```python
# PT1 = Bottom right corner
cv2.rectangle(cblank, pt1=(600,600), pt2=(800,800), color=(0,255,0), thickness=5)
plt.imshow(cblank)
```

```python
# Drawing a circle
# Center, Radius (Width, Height)

cv2.circle(cblank, center=(300,300), radius=100, color=(255,255,255), thickness=10)
plt.imshow(cblank)
```

```python
# Drawing a circle
# Center, Radius (Width, Height)

cv2.circle(cblank, center=(700,700), radius=100, color=(225,100,150), thickness=10)
plt.imshow(cblank)
```

```python
# Drawing a circle
# Center, Radius (Width, Height)

cv2.circle(cblank, center=(700,100), radius=50, color=(225,150,150), thickness=-1)
plt.imshow(cblank)
```
