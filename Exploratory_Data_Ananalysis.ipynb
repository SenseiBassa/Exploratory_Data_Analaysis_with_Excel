BIVARATE ANALYSIS

```python
sb.jointplot(x='horsepower', y='price', data=df)
```

- We can observe a positive correlation between the horsepower feature and our dependent variable 
- The horsepower feature is fairly symmetrical and the presence of some outliers were detected.

```python
sb.scatterplot(x='horsepower', y='price', data=df)
```

- We can observe a positive correlation between the horsepower feature and our dependent variable 

```python
# relationship between the cars engine and the car price
```

```python
sb.boxplot(x='enginelocation', y='price', data=df)
```

- We can extrapolate from the figure above that on average, less expensive cars have their engines in front of the car though we detected the presence of outliers
- Additionally, all cars that have their engine location at the rear are on average more expensive than those that have engines in front.

```python
sb.barplot(x='enginelocation', y='enginesize', data=df)
```

- We could deduce that rear engine cars tend to be larger in size than front enine cars.
- Therefore, sizes plays a role in the location of the engine.

```python
 var = ['symboling', 'wheelbase', 'carlength', 'carwidth', 'stroke', 'citympg', 'compressionratio', 'price']
 sb.pairplot(df[var])
```

```python
plt.figure(figsize=(10,7))
sb.heatmap(df[var].corr(),annot=True)
```

```python
g = sb.pairplot(df[var], kind='reg', plot_kws={'line_kws': {'color': 'red'}, 'scatter_kws': {'alpha': 0.4}})
g.fig.set_size_inches(12, 12)
g.fig.suptitle('Multivariate analysis')
```
